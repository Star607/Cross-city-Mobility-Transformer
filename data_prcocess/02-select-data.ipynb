```python
data = 'New York'
city = data.replace(" ", "_")
city
```

```python
df = pd.read_csv(f'./rawData/Foursquare_global/city_user_day_fix/res/{city}.csv')
df.head()
```

```python
# val, cnt = np.unique(df['traj_id'].to_numpy(), return_counts=True)
traj_id_f = np.unique(cnt[0][np.where(cnt[1]>=6)[0]])
df_f = df[df['traj_id'].isin(traj_id_f)]
df_f.head()
```

```python
df_f = df_f.sort_values(by = ['user', 'day', 'check-in_time']).reset_index(drop=True)
pois = np.unique(df_f['location_id'])
pois_id_dict = dict(zip(pois, range(len(pois))))
df_f['poi_id'] = df_f['location_id'].apply(lambda x:pois_id_dict[x])
Path('./rawData/Foursquare_global/city_user_day_fix/cities/').mkdir(parents=True, exist_ok=True)
df_f.to_csv(f'./rawData/Foursquare_global/city_user_day_fix/cities/{city}_f.csv', index=False)
grouped = df_f.groupby(['user', 'day'])
u_l, seq_l, t_l = [], [], []
for name, g in grouped:
    u_l.append(g['user'].values.tolist()[0])
    seq_l.append(g['poi_id'].values.tolist())
    t_l.append(g['check-in_time'].values.tolist())
len(u_l), len(seq_l), len(seq_l)
```

```python
m_tes_tra_ratio, m_tes_val_ratio, m_tes_tes_ratio = 0.7, 0.1, 0.2
all_idx = [i for i in range(len(seq_l))]
np.random.shuffle(all_idx)
if city in TEST_CITY:
    tra_idx = all_idx[ : int(len(seq_l) * m_tes_tra_ratio)]
    tra_user = [u_l[id] for id in tra_idx]
    tra_path = [seq_l[id] for id in tra_idx]
    tra_tim = [t_l[id] for id in tra_idx]
    
    val_idx = all_idx[int(len(seq_l) * m_tes_tra_ratio) : int(len(seq_l) * (m_tes_tra_ratio + m_tes_val_ratio))]
    val_user = [u_l[id] for id in val_idx]
    val_path = [seq_l[id] for id in val_idx] 
    val_tim = [t_l[id] for id in val_idx]

    tes_idx = all_idx[int(len(seq_l) * (m_tes_tra_ratio + m_tes_val_ratio)):] 
    tes_user = [u_l[id] for id in tes_idx]
    tes_path = [seq_l[id] for id in tes_idx]
    tes_tim = [t_l[id] for id in tes_idx]
    
data = {'tra_user': tra_user, 'tra_path': tra_path, 'tra_tim': tra_tim, 
        'val_user': val_user, 'val_path': val_path, 'val_tim': val_tim,
        'tes_user': tes_user, 'tes_path': tes_user, 'tes_tim': tes_user}
data_str = json.dumps(data)
with open(f'./rawData/Foursquare_global/city_user_day_fix/cities/{city}_split.json', 'w') as file:
    file.write(data_str)
```
